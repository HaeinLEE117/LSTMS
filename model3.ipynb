```python
import json
import string
import os
import numpy as np
from numpy import array
from keras.models import Sequential
from keras.layers import Dense, LSTM, Bidirectional,Input
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

    
```

```python
evaluating = {
    'best':1,
    'bad':0,
    'normal':0
}
```

# data 구성

```python

ann_target = 'C:/Users/haeinlee/Desktop/train_data/'
data_flag = 0

for (root_path, dir, files) in os.walk(ann_target):
    for d in dir:
        # 각 디렉토리에서 8개의 파일을 가져와야 함
        for (path, dir, files) in os.walk(root_path+'/'+d+'/'):
            j_index = 0
            directory_data=[]
            for file in files:
                try:
                    with open(path+file, "r", encoding='UTF8') as json_file:
                        nj_data = json.load(json_file)
                        print(type(nj_data))
                        if type(nj_data) =='list':
                            nj_data = nj_data[0]
                except:
                    pass
                try:
                    tmp = np.array(sum(nj_data['keypoints'],[]))
                    tmp[0::2] = tmp[0::2]/nj_data['resolution'][0]
                    tmp[1::2] = tmp[1::2]/nj_data['resolution'][1]
                    directory_data.append(tmp)
                    j_index+=1
                except:
                    print("json Error: "+file)

            if len(directory_data)<8:
                print("failed : "+path)
                
            else:
                if data_flag==0:
                    x=np.array([directory_data])
                    y=np.array([evaluating[nj_data['evaluation']]])
                    data_flag=1
                    print(x.shape)
                else:
                    x= np.concatenate((x, np.array([directory_data])),axis=0)
                    y= np.concatenate((y, [evaluating[nj_data['evaluation']]]))
```

```python
x.shape
```

```python
# 데이터를 랜덤하게 섞고 트레인, 테스트 셋을 나눠줍시다...
x_train, x_test, y_train, y_test = train_test_split(x, y,  test_size=0.2, random_state=42)
```

```python
np.unique(y,return_counts=True) 
```

#LSTM 모델

```python
 
model3 = Sequential()

forward_layer = LSTM(20, activation = 'sigmoid',return_sequences=True)
backward_layer = LSTM(20, return_sequences=True, go_backwards=True)

# model.add(Input(shape = (8,24)))
model3.add(Bidirectional(forward_layer,backward_layer=backward_layer, input_shape=(8, 24)))
model3.add(Dense(8))
model3.add(Dense(1, activation = 'sigmoid'))
 
model3.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
 
# from keras.callbacks import EarlyStopping
# early_stopping = EarlyStopping(monitor='binary_crossentropy', patience=10, mode='auto',)
# mode=auto loss면 최저값이10번정도 반복되면 정지, acc면 최고값이 10번정도 반복되면 정지
# mode=min, mode=max
# history = model.fit(x_train, y_train, epochs=1000, batch_size=10, verbose=2, callbacks=[early_stopping])

history = model3.fit(x_train, y_train, epochs=300, batch_size=40, verbose=1,validation_split=0.3)
```

```python
history.history.keys()
```

테스트 데이터셋 떼어놓은거 검증

```python
results = model3.evaluate(x_test, y_test, batch_size=10)
print("test loss, test acc:", results)
```

```python

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```
